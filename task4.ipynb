**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(img_in):
    result = numpy.zeros(img_in.shape)
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum =[]
        if p[0] > 0 and p[1] > 0 and p[0] < img_in.shape[0]-1 and p[1] < img_in.shape [1]-1:
            for q in np.ndindex(3,3):
                if q[1] == 0:
                    neighborhood_sum.append(img_in[p[0]+q[0]-1, p[1]+q[1]-1]*1)
                if q[1] ==2:
                    neighborhood_sum.append(img_in[p[0]+q[0]-1, p[1]+q[1]-1]*(-1))
            result[p] = sum(neighborhood_sum)//6
        else:
            result[p] = 0
    return result
def prewitt_v(img_in):
    result = numpy.zeros(img_in.shape)
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum =[]
        if p[0] > 0 and p[1] > 0 and p[0] < img_in.shape[0]-1 and p[1] < img_in.shape [1]-1:
            for q in np.ndindex(3,3):
                if q[0] == 0:
                    neighborhood_sum.append(img_in[p[0]+q[0]-1, p[1]+q[1]-1]*1)
                if q[0] ==2:
                    neighborhood_sum.append(img_in[p[0]+q[0]-1, p[1]+q[1]-1]*(-1))
            result[p] = sum(neighborhood_sum)//6
        else:
            result[p] = 0
    return result
```
